```python
from scipy.stats import wilcoxon

our_method_aurocs = [0.7110, 0.7069, 0.7240, 0.7601, 0.6699]
cnnlstm_method_aurocs = [0.6939, 0.6540, 0.6758, 0.7403, 0.7647]

stat, p = wilcoxon(our_method_aurocs, cnnlstm_method_aurocs, alternative='greater')
print(f"Wilcoxon signed-rank test statistic: {stat}, p-value: {p}")
```

```python
from scipy.stats import ttest_rel

stat, p = ttest_rel(our_method_aurocs, cnnlstm_method_aurocs)
print(f"Paired t-test statistic: {stat}, p-value: {p}")
```

```python
import numpy as np

diffs = np.array(our_method_aurocs) - np.array(cnnlstm_method_aurocs)

boot = []
for _ in range(10000):
    sample = np.random.choice(diffs, size=len(diffs), replace=True)
    boot.append(np.mean(sample))

ci = np.percentile(boot, [2.5, 97.5])
print("Mean diff:", np.mean(diffs))
print("95% CI:", ci)
```
