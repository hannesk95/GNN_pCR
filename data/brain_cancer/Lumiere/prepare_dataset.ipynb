```python
import os
from glob import glob
from tqdm import tqdm
import torchio as tio
```

```python
output_dir = "/home/johannes/Data/SSD_2.0TB/GNN_pCR/data/brain_cancer/Lumiere/processed"

for file in tqdm(sorted(glob("/home/johannes/Data/SSD_2.0TB/GNN_pCR/data/brain_cancer/Lumiere/raw/Patient*/week*/HD-GLIO*/reg*/*reg.nii*"))):

    patient_id = file.split("/")[-5]
    date = file.split("/")[-4]

    os.makedirs(os.path.join(output_dir, patient_id, date), exist_ok=True)

    img = tio.ScalarImage(file)
    img_aligned = tio.ToCanonical()(img)
    img_resampled = tio.Resample((1, 1, 1))(img_aligned)
    img_normalized = tio.ZNormalization()(img_resampled)
    output_path = os.path.join(output_dir, patient_id, date, os.path.basename(file))

    if not "gz" in output_path:
        output_path += ".gz"

    img_normalized.save(output_path)
```
